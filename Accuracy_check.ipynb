```python
import files2rouge
def checkScore(ref_file,hyp_file,i):
    print("processing batch no: ", i)
    return files2rouge.run(hyp_file, ref_file)
```

```python
lines_per_file = 500
smallfile = None
noof_ref_files=0
noof_hyp_files=0
with open('references.txt') as bigfile:
    for lineno, line in enumerate(bigfile):
        if lineno % lines_per_file == 0:
            if smallfile:
                noof_ref_files+=1
                smallfile.close()
            small_filename = 'reference_{}.txt'.format(noof_ref_files)
            smallfile = open(small_filename, "w")
        smallfile.write(line)
    if smallfile:
        smallfile.close()

smallfile = None
        
with open('summaries.txt') as bigfile:
    for lineno, line in enumerate(bigfile):
        if lineno % lines_per_file == 0:
            if smallfile:
                noof_hyp_files+=1
                smallfile.close()
            small_filename = 'summary_{}.txt'.format(noof_hyp_files)
            smallfile = open(small_filename, "w")
        smallfile.write(line)
    if smallfile:
        smallfile.close()
```

```python
all_score=list()
for i in range(1,noof_hyp_files):
    ref_file='reference_{}.txt'.format(i)
    hyp_file='summary_{}.txt'.format(i)
    all_score.append(checkScore(ref_file,hyp_file,i))
    print(ref_file,hyp_file)
```
